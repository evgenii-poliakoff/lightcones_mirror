```python
import math
import tools
from scipy.linalg import eigh, eigvalsh
from scipy.sparse import coo_matrix
```

```python
# check for lambda[i]
n_spin = 1
m_spin = tools.spin_chain(n_spin)

m = tools.spin_chain_boson_model(n_spin, 5, 5)
lo = m.get_local_observables()

lambd = np.arange(0.05, 2, 0.05)
J = 1
r = 0.5
h = J/lambd

lambda_index = 19
h_lambda = J/lambd[lambda_index]

H_spin_ten = coo_matrix((m.dimension , m.dimension ), dtype = complex).tocsc()
for i in range(n_spin - 1):
    H_spin_ten -= J * (
        (1 + r)/2 * m.sx[i] @ m.sx[i+1] +
        (1 - r)/2 * m.sy[i] @ m.sy[i+1])
# periodic boundary condition    
H_spin_ten -= J * (
    (1 + r)/2 * m.sx[n_spin - 1] @ m.sx[0] +
    (1 - r)/2 * m.sy[n_spin - 1] @ m.sy[0])
for i in range(n_spin):
    H_spin_ten -= h_lambda * m.sz[i]


H_spi = np.zeros((2**n_spin, 2**n_spin), dtype='complex')
for i in range(n_spin - 1):
    H_spi -= J * ((1 + r)/2 * m_spin.sx[i] @ m_spin.sx[i+1] +
                (1 - r)/2 * m_spin.sy[i] @ m_spin.sy[i+1])
H_spi -= J * ((1 + r)/2 * m_spin.sx[n_spin - 1] @ m_spin.sx[0] +
                (1 - r)/2 * m_spin.sy[n_spin - 1] @ m_spin.sy[0])
for i in range(n_spin):
    H_spi -= h_lambda * m_spin.sz[i]

eig, eiv = eigh(H_spi)

# bosons vacuum
psi_boson_vac = np.zeros(m.fs_chain.dimension)
psi_boson_vac[0] = 1.0  

# total initial condition
psi_ini_eig = np.kron(eiv[:,0], psi_boson_vac)

# check tensor product
dim_boson = m.fs_chain.dimension
dim_spin = m.space.dimension
if np.allclose(H_spin_ten.toarray(), np.kron(H_spi, np.eye(dim_boson))) == True:
    print('Tensor product works correctly')

# # check that ini condition is eigenstate 
# print('Psi_ini is eigenstate H_spin', np.linalg.norm(H_spin_ten.dot(psi_ini_eig) - eig[0] * psi_ini_eig))
```

```python
# проверка работы частичного следа
rho_full = np.outer(psi_ini_eig, np.conj(psi_ini_eig))
rho_full = rho_full.reshape((dim_spin, dim_boson, dim_spin, dim_boson))
rho_sys = np.einsum('iaja->ij', rho_full)

np.allclose(rho_sys, np.outer(eiv[:,0], np.conj(eiv[:,0])))
```

```python
# проверка разложения Шмидта1
psi_eig_after, j_prob_eig, *_  = lo.quantum_jumpEx(psi_ini_eig, 1)

rho_full = np.outer(psi_eig_after, np.conj(psi_eig_after))
rho_full = rho_full.reshape((dim_spin, dim_boson, dim_spin, dim_boson))
rho_sys = np.einsum('iaja->ij', rho_full)

print(j_prob_eig)
np.allclose(rho_sys,np.outer(eiv[:,0], np.conj(eiv[:,0])))
```

```python
# проверка разложения Шмидта2
psi_spin = np.zeros(2**n_spin) 
psi_spin[1] = 1.0
psi_ini = np.kron(psi_spin, psi_boson_vac)

psi_after, j_prob_eig, *_  = lo.quantum_jumpEx(psi_ini, 1)

rho_full = np.outer(psi_after, np.conj(psi_after))
rho_full = rho_full.reshape((dim_spin, dim_boson, dim_spin, dim_boson))
rho_sys = np.einsum('iaja->ij', rho_full)

print(j_prob_eig)
print(np.allclose(rho_sys,np.outer(psi_spin, np.conj(psi_spin))))
```

```python
import numpy as np
import tools
from tools import mv
from evolution import evolution
import time

tmax = 500
dt = 0.01
t = np.arange(start = 0, stop = tmax, step = dt)
nt = t.size

ns = len(e) # number of chain sites we keep

H1 = tools.tridiag(e[:], h[1:]) # Hamiltonian for free propagation of a single quantum

phi_ini = np.zeros(ns, dtype = complex)
phi_ini[0] = 1 # initially quantum is on the site 0

phi_lc = np.zeros((ns, nt), dtype = np.cdouble) # Here we store the propagated orbitals

def apply_H1t(ti, phi_in, phi_out):
    mv(H1, phi_in, phi_out, cout=1)
    return

def eval_Olc(ti, phi):
    phi_lc[:, ti] = phi

phi = np.zeros(ns, dtype = complex)
phi_mid = np.zeros(ns, dtype = complex)
phi_mid_next = np.zeros(ns, dtype = complex)
    
start_time = time.time()
    
evolution(0, nt, dt, apply_H1t, eval_Olc, phi_ini,  phi, phi_mid, phi_mid_next)

end_time = time.time()
    
print('execution time: ', end_time-start_time, ' sec')
```

```python
import os
import subprocess
import shlex
import time
import numpy as np

lambd = np.arange(0.05,2,0.1)
nl = len(lambd)
Nspin = 5
Nspin_ = str(Nspin)

jobs = {}

maxcores = 30

def wait_running_less_than(n):
    global jobs
    while len(jobs) >= n:
        finished = {k: v for k, v in jobs.items() if v[0].poll() is not None}
        for k, v in finished.items():
            if v[0].returncode != 0:
                print(f"Job {k} failed with return code {v[0].returncode}")
            v[1].close()
        jobs = {k: v for k, v in jobs.items() if k not in finished}
        time.sleep(0.5)


for i in range(nl):
    wait_running_less_than(maxcores)

    i_ = str(i)
    lambd_ = lambd[i]
    
    # Создание папки fout для каждого начального условия
    fout = os.path.join('проверка', f'lambd_{i_}')
    os.makedirs(fout, exist_ok=True)  

    cmd = (f'python impurity_solver_spin_chain.py --lambd={lambd_} --nspins=5 '
           f'--cfrom=1 --cto=1 --csize=1 --fout={fout} --nquanta=5 '
           f'--tmax=5 '
           f'--intervals={intervals_filename} --couplings={couplings_filename} '
           f'--rotations={rotations_filename} --times={times_filename} '
           f'--star_out={star_out_filename}')

    log_path = os.path.join(fout, "log.out")
    with open(log_path, 'w') as logf:
        p = subprocess.Popen(shlex.split(cmd), shell=False, stdout=logf, stderr=logf)
        jobs[(i)] = (p, logf)

wait_running_less_than(1)

print('Done!')
```
